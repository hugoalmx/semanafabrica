#2. Plotando Gráficos

```python
#Plotando gráficos
#2A = Funcionário em cada setor

numero_funcionarios_por_departamento = df.groupby(['Department'])[['EmployeeCount']].count()
numero_funcionarios_por_departamento


numero_funcionarios_por_departamento.plot(kind = 'bar', color='green')
plt.title('Quantidade de Funcionários por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Número de Funcionários')
```

```python
#Plotando gráficos
#2B = Quantidade de Funcionários para cada nível de escolaridade

numero_funcionarios_por_escolaridade = df.groupby(['Education'])[['EmployeeCount']].count()
numero_funcionarios_por_escolaridade

numero_funcionarios_por_escolaridade.plot(kind = 'line', color='blue', marker = 'v')
plt.title('Quantidade de Funcionários por Nível de Escolaridade')
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Número de Funcionários')
```

```python
#Plotando gráficos
#2C = Quantidade de Funcionários que tem ou não algum atrito na empresa

numero_funcionarios_atritos = df.groupby(['Attrition'])[['EmployeeCount']].count()
numero_funcionarios_atritos

numero_funcionarios_atritos.plot(kind = 'pie', subplots=True, colors = ['gold', 'lightgreen'], autopct='%1.1f%%')
plt.legend(loc='best')
plt.title('Taxa de Atrito na Empresa')

```
